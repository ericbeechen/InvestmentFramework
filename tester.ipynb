```python
def qlike_loss(realized_var, pred_var):
    """
    QLIKE formula: (r^2 / sigma^2) - log(r^2 / sigma^2) - 1
    """
    eps = 1e-10
    # Ensure predicted variance is positive to avoid division by zero or log of negative
    pred_safe = np.maximum(pred_var, eps)
    ratio = realized_var / pred_safe
    return ratio - np.log(ratio + eps) - 1

def qlike_loss_gradients(realized_var, pred_var):
    """
    Gradients for QLIKE:
    dL/d_realized = 1/pred - 1/realized
    dL/d_pred     = (pred - realized) / (pred^2)
    """
    eps = 1e-10
    pred_safe = np.maximum(pred_var, eps)
    real_safe = np.maximum(realized_var, eps)
    
    dL_drealized = (1.0 / pred_safe) - (1.0 / real_safe)
    dL_dpred = (pred_safe - realized_var) / (pred_safe**2)
    
    return dL_drealized, dL_dpred

def mse_loss(realized_var, pred_var):
    """
    Weighted MSE formula: sigma^2 * ((r^2 / sigma^2) - 1)^2
    Simplified this is equivalent to: (realized - pred)^2 / pred
    """
    eps = 1e-10
    pred_safe = np.maximum(pred_var, eps)
    ratio = realized_var / pred_safe
    return pred_safe * (ratio - 1)**2

def mse_loss_gradients(realized_var, pred_var):
    """
    Gradients for the specific weighted MSE formula provided:
    L = (x - y)^2 / y  where x=realized, y=pred
    dL/dx = 2(x - y) / y
    dL/dy = (y^2 - x^2) / y^2 = 1 - (x/y)^2
    """
    eps = 1e-10
    pred_safe = np.maximum(pred_var, eps)
    
    # dL/drealized
    dL_drealized = 2 * (realized_var - pred_safe) / pred_safe
    
    # dL/dpred: Derived via quotient rule on (x-y)^2 / y
    ratio = realized_var / pred_safe
    dL_dpred = 1.0 - (ratio**2)
    
    return dL_drealized, dL_dpred
```

```python
def procedure_5_1(cov_matrices, returns, loss_func, epsilon_tol=1e-6, max_iter=5000):
    """
    Random Portfolios Average Variance Testing.
    
    Args:
        cov_matrices: Array/List of candidate covariance matrices (T x N x N)
        returns: Array of realized returns (T x N)
        loss_func: Function handle for loss calculation (e.g., qlike_loss)
        epsilon_tol: Tolerance level for stopping criterion
        max_iter: Maximum iterations safeguard
    """
    T = len(returns)
    N = returns.shape[1]
    
    L_tot = 0.0
    n_iter = 0
    
    for _ in range(max_iter):
        # Step 3: Generate random weight vector w ~ N(0, I), then normalize
        w = np.random.standard_normal(N)
        w = w / np.linalg.norm(w)
        
        # Choose s uniformly at random from 1 to T
        s = np.random.randint(0, T)
        
        # Calculate realized variance for the portfolio: (r_s^T * w)^2
        realized_var = (returns[s] @ w)**2
        
        # Calculate predicted variance for the portfolio: w^T * Omega_s * w
        pred_var = w.T @ cov_matrices[s] @ w
        
        # Step 4: Compute loss L for this specific instance
        current_loss = loss_func(realized_var, pred_var)
        
        L_tot += current_loss
        n_iter += 1
        
        # This means we BREAK if L[...] / L_tot < epsilon_tol
        if L_tot > 0 and (current_loss / L_tot) < epsilon_tol:
            break
            
    if n_iter == 0:
        return np.nan
            
    # Step 6: Output Average Loss
    return L_tot / n_iter
```

```python
def procedure_5_2(cov_matrices, returns, loss_func, loss_gradients, epsilon_tol=1e-6, max_iter=5000):
    """
    Random Portfolios Average Variance Testing with gradient updates on w.
    
    Args:
        cov_matrices: Array/List of candidate covariance matrices (T x N x N)
        returns: Array of realized returns (T x N)
        loss_func: Function handle for loss calculation (e.g., qlike_loss)
        loss_gradients: Function (realized_var, pred_var) -> (dL_drealized, dL_dpred)
        epsilon_tol: Tolerance level for stopping criterion (gradient norm)
        max_iter: Maximum iterations safeguard
    """
    T = len(returns)
    N = returns.shape[1]
    
    n_iter = 0
    avg_loss = 0.0
    
    for _ in range(max_iter):
        # Step 3: Generate random weight vector w ~ N(0, I), then normalize
        w = np.random.standard_normal(N)
        w = w / np.linalg.norm(w)
        
        # Choose s uniformly at random from 1 to T
        s = np.random.randint(0, T)
        r_s = returns[s]
        omega_s = cov_matrices[s]
        
        # Calculate realized variance for the portfolio: (r_s^T * w)^2
        realized_var = (r_s @ w)**2
        
        # Calculate predicted variance for the portfolio: w^T * Omega_s * w
        pred_var = w.T @ omega_s @ w
        
        # Step 4: Compute loss L for this specific instance
        current_loss = loss_func(realized_var, pred_var)
        dL_drealized, dL_dpred = loss_gradients(realized_var, pred_var)
        
        # Gradient w.r.t. w via chain rule
        grad_realized = 2.0 * (r_s @ w) * r_s
        grad_pred = 2.0 * (omega_s @ w)
        grad_w = dL_drealized * grad_realized + dL_dpred * grad_pred
        
        # Update w with step size 1 / n_iter
        n_iter += 1
        w = w - (1.0 / n_iter) * grad_w
        
        # Online average loss (no L_tot accumulator)
        avg_loss += (current_loss - avg_loss) / n_iter
        
        if np.linalg.norm(grad_w) < epsilon_tol:
            break
            
    if n_iter == 0:
        return np.nan
            
    return avg_loss
```

```python
# Example implementation logic for Procedure 5.2 Step 4
def step4_update(returns_s, Omega_s, w, loss_gradients, n_iter):
    """
    Single gradient update for Step 4 using the provided loss gradients.
    """
    realized_var = (returns_s @ w) ** 2
    pred_var = w.T @ Omega_s @ w
    
    dL_dx, dL_dy = loss_gradients(realized_var, pred_var)
    
    # Gradient of realized variance w.r.t w: d/dw (r^T w)^2 = 2 * (r^T w) * r
    grad_x = 2 * (returns_s @ w) * returns_s
    
    # Gradient of predicted variance w.r.t w: d/dw (w^T Omega w) = 2 * Omega * w
    grad_y = 2 * (Omega_s @ w)
    
    # Total gradient for the weight update
    nabla_w = dL_dx * grad_x + dL_dy * grad_y
    
    w = w - (1 / n_iter) * nabla_w
    return w, realized_var, pred_var
```
